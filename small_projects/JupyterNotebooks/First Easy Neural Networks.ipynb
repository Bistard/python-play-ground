```python
input = 40
weight = 0.3
goal_pred = 100
alpha = 0.001
for i in range(100):
    pred = input * weight
    delta = pred - goal_pred
    error = delta ** 2
    derivative = input * delta
    weight = weight - derivative * alpha
    print('error:' + str(error), 'prediction:' + str(pred), 'weight:' + str(weight))
```
