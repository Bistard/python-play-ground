```python
def nerual_networks(input, weight):
    pred = w_sum(input, weight)
    return pred


def w_sum(a, b):
    assert len(a) == len(b)
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]
    return output


def ele_mul(num, vect):
    output = [0, 0, 0]
    assert len(output) == len(vect)
    for i in range(len(vect)):
        output[i] = num * vect[i]
    return output


toes =  [8.5 , 9.5, 9.9, 9.0] 
wlrec = [0.65, 0.8, 0.8, 0.9] 
nfans = [1.2 , 1.3, 0.5, 1.0]
weights = [0.1, 0.2, -.1] 
win_or_lose_binary = [1, 1, 0, 1]
true = win_or_lose_binary[0]

alpha = 0.01

for i in range(30):
    input = [toes[0],wlrec[0],nfans[0]]
    pred = nerual_networks(input, weights)
    delta = pred - true
    error = (delta) ** 2
    weight_delta = ele_mul(delta, input)


    for i in range(len(weight_delta)):
        weights[i] -= weight_delta[i] * alpha
    print('------------'+'\n''error:'+str(error), 'predictions:'+str(pred))
    
```
